```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('Diwali Sales Data.csv', encoding= 'unicode_escape')
```

```python
df.info()
```

```python
df.shape
```

```python
df.head()
```

```python
df.drop(['Status', 'unnamed1'], axis=1, inplace=True) #axis means the whole row
#inplace is used to save the command we have passed.
```

```python
pd.isnull(df).sum() #.sum()gives us sum of all the columns
```

```python
df.dropna(inplace=True)
```

```python
df['Amount'] = df['Amount'].astype('int') #this will change the float values present in data types in integer values respectivrly
```

```python
df['Amount'].dtypes
```

```python
df.columns
```

```python
df.rename(columns={'Marital_Status': 'Shaadi'}) #here the columns name in dictionary format because of this reason here we have take the curly brackets respectively.
```

```python
df.describe()
```

```python
df[['Age','Orders','Amount']].describe()
```

```python
ax = sns.countplot(x = 'Gender', data = df)
```

```python
ax = sns.countplot(x= 'State', data =df)
```

```python
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x = 'Gender',y= 'Amount', data =sales_gen)
```

from above graph we caN understand that most of the buyers are the female. even the purchasing power of female is greater than mens respectivrly

```python
ax = sns.countplot(data = df, x = 'Age Group', hue = 'Gender') #hue is used to show as a scale for the given graph here we used blue colrs for mens and yellow colors for the women respectively

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x = 'Age Group',y= 'Amount' ,data = sales_age)
```

from above graph we can see the most of the buyer are betweeb age group 26-35

```python
ax = sns.countplot(data = df, x = 'Marital_Status')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)


sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(data = sales_state, x = 'Marital_Status',y= 'Amount', hue='Gender')
```

```python

sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,7)})
sns.barplot(data = sales_state, x = 'State',y= 'Orders')
```

```python
sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,7)})
sns.barplot(data=sales_state, x='State', y='Amount')
plt.show()
```
